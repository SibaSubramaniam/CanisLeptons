```python
def sharp_ratio(df,risk=0):
    close=df['Close']
    f_close=close.shift(-1)
    
    labs=df['label']
    ret_=((f_close/close)-1).dropna()
    
    ret_s=ret_.shift(-1)
    sharp_ratio=((ret_s*labs)-risk)/(ret_s*labs).std()
    
    df['return']=ret_
    df['sharp_ratio']=sharp_ratio
    
    return df
```
