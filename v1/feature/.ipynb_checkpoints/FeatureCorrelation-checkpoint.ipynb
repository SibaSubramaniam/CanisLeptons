# Experiment Region
This is where you can set different periods for 5 sma or 5 ema features and get their correlation

```python
def get_five_fdf_corr_stat(df,step=5,mode='sma',dropna=True):
    """
    args:
        df : dataframe with date,o,h,l,c
        mode : 'sma' or 'ema'
    return:
        returns correlation heatmap between 5 features
    """
    five_df = fe_ob.get_five_features(df,step,mode,dropna)
    # print(five_df.columns)
    corr_five = five_df.corr()
    
    return five_df,fesv_ob.corr_heatmap(corr_five)

five_df, five_heat = get_five_fdf_corr_stat(MAIN_df,6,mode='ema')

five_heat
```
