# Real Time Features

Calculate features for realtime data and predict labels for the same

Now write a code that takes each line as input and generate labels using a pre trained machine learning model. 


##  `Time(in milli sec, unix time), Open, High, Low, Close, Price, Volume`

```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
from tqdm import tqdm
```

```python
from realTime import RealTimeExt
re_ob = RealTimeExt()

from loadData import LoadData
ld_data = LoadData()
```

# Load Trained model

```python
rf_model = joblib.load('../trained_randomForest.sav')
```

```python
pred_y = []
MAX_MEMORY = 150
REMOVE_COLS = 50

with open('../../data/real_time_data/stockPriceRecord.txt') as f:
    cols = ['Date','Open', 'High', 'Low', 'Close', 'Price', 'Volume']
    counter = 0 
    cal_df = pd.DataFrame(columns=cols)
    
    while(f):
        try:
            print("i = ",counter,end=' ')
            line = f.readline()
            l = line.split(',')
            l = l[:7]  # get required columns
            
            # get DataFrame
            t = re_ob.process_list(l,counter) 
            
            # calculate features for cal_df
            cal_df = cal_df.append(t,ignore_index=True,sort=True)
            
            
                
            try:
                ext_vals = re_ob.get_features_real_time(cal_df)
                print(ext_vals)
                # predict using machine learning model here
                ext_vals = [ext_vals]
                t = rf_model.predict(ext_vals)
                pred_y.append(t[0])
                print("label : ",t)
            
            except Exception as e:
                # print(e)
                pass
            # returns empty df if period not met
           
        
            if (counter%MAX_MEMORY == 0 and counter!=0):
                cal_df = cal_df.iloc[REMOVE_COLS:]
                print(cal_df.shape)
                
                
            counter+=1            
            
        except Exception as e:
            # print(e)
            print("End of File")
            break
            
        
```
